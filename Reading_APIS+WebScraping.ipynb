# WebScraping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
web_page = requests.get('https://www.ambitionbox.com/list-of-companies?page=1').text
```

```python
soup = BeautifulSoup(web_page, 'lxml')
```

```python
print(soup.prettify())
```

```python
company = soup.find_all('div', class_='company-content-wrapper')
```

```python
names=[]
ratings=[]
company_types = []
for i in company:
  name = i.find('h2', class_='company-name bold-title-l')
  rating = i.find('p', class_ = 'rating').text.strip()
  company_type = i.find_all('p',class_='infoEntity sbold-list-header')[0].text.strip()
  names.append(name)
  ratings.append(rating)
  company_types.append(company_type)
```
