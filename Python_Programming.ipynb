# Object Oriented Programming

## I. Class
Data or Property<br>
Function or Behaviour

```python
class Car:
  color=None
  top_speed=None

  def __init__(self):
    print('Welcome')

  def get_info(self, x, y):
    self.color=x
    self.top_speed=y
    print('You have a '+self.color+' car with the top speed of '+str(self.top_speed))

  def cal_Avg_speed(self,km,t):
    speed=km/t
    print(speed)
```

```python
car = Car()
car.get_info('pink',120)
car.cal_Avg_speed(120,6)
```

```python
class Fraction:
  numerator = None
  denominator = None

  def __init__(self,x,y):
    self.numerator = x
    self.denominator = y

  def __str__(self):
    return '{0}/{1}'.format(self.numerator,self.denominator)
```

```python
frac = Fraction(3,4)
```

```python
print(frac)
```

```python
char=[]
name='Manish Sharma'
characters = list(set([i for i in name]))
dictionary = {}
for i in characters:
  dictionary[i]=[j for j in name].count(i)
for i in dictionary.keys():
  if(dictionary[i]>1):
    char.append(i)
```
